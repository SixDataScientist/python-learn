# Conda Environment Management Commands

## Creating and Managing Environments
```bash
# Create a new environment with Python 3.10
conda create --name py310 python=3.10  

# Activate the environment
conda activate py310  

# Deactivate the current environment
conda deactivate  

# List all conda environments
conda env list    

# Rename an environment
conda rename -n py310 test_env  

# Remove an environment
conda env remove --name py310  
```

## Package Management
```bash
# Install a specific Python version
conda install python=3.8   

# Update numpy
conda update numpy  

# Remove numpy
conda remove numpy  

# Install a specific numpy version using pip
pip install numpy==1.25   
```

## Environment Export and Import
```bash
# Export current environment to a YAML file
conda env export > environment.yml   

# View the contents of the environment file
cat environment.yml  

# Creating an Environment from a File
conda env create -f environment.yml   
```

## Using Keyboard Shortcuts in VS Code  

### Region Control
- **Collapse the Current Region:**  
  > Press Ctrl + Shift + [  

- **Expand the Current Region:**  
  Place the cursor inside the collapsed block.  
  > Press Ctrl + Shift + ]  

- **Expand the Current Region and Its Subregions:**  
  > Windows/Linux: Ctrl + K, then Ctrl + ]  

- **Collapse All Regions:**  
  > Press Ctrl + K, then Ctrl + 0  

- **Expand All Regions:**  
  > Press Ctrl + K, then Ctrl + J  

### Line Operations
- **Copy Line Up/Down:** Duplicate lines above or below.  
  > Windows/Linux: Alt + Shift + ↑ / ↓  

- **Delete Line:** Remove the current line.  
  > Windows/Linux: Ctrl + Shift + K  

### Comments
- **Block Comment:** Toggle block comments.  
  > Windows/Linux: Alt + Shift + A
